```python
# importing required libraries in Python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from sklearn import tree
warnings.filterwarnings(action= 'ignore')

%matplotlib inline
```

```python
# loading dataset

from sklearn import datasets

data= datasets.load_iris()
X = pd.DataFrame(data.data, columns = data.feature_names)
X.head()
```

```python
Y= data.target
Y
```

```python
# splitting train and test dataset

from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test= tts(X, Y, train_size= 0.80, stratify= Y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# defining Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

### So, we created the Decision Tree Classifier above and next we will visualize it.

```python
# visualising the graph

plt.figure(figsize=(15,10), dpi= 120)
tree.plot_tree(dtc, feature_names= data.feature_names, class_names= data.target_names, filled= True)
```

### Now, let's evaluate our model.

```python
# observe evaluation metrics for Train set using Classification Report

from sklearn.metrics import classification_report
predicted_values= dtc.predict(x_train)
print(classification_report(y_train, predicted_values))
```

```python
# observe accuracy score for Train set

from sklearn.metrics import accuracy_score
accuracy_score(y_train, predicted_values)
```

```python
# observe evaluation metrics for Test set using Classification Report

from sklearn.metrics import classification_report
Predicted_values= dtc.predict(x_test)
print(classification_report(y_test, Predicted_values))
```

```python
# observe accuracy score for Test set

accuracy_score(y_test, Predicted_values)
```

### Higher accuracy implies more accurate model!

### Here, we observe that the difference between the test and train accuracy scores is very less and thus, no issue of overfitting .

### After this we can feed any new data and the model can predict the correct class !
